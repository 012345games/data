2.Get list of all categories from main page

```python
Dpages=[]
Dcells=[]
Dall_cat_links=[]
url_D='https://www.gameflare.com/3d-games/'
Dresponse=requests.get(url_D)
Dcat_page=BeautifulSoup(Dresponse.text,'html.parser')
Dpages=Dcat_page.find('ul',class_='nav nav-sidebar')
Dlinks=Dpages.find_all('a',href=True)
for Dlink in Dlinks:
    Dgame=Dlink['href']
    if '/videos/' in Dgame:
        pass
    else:
        Dall_cat_links.append(Dgame)

Dall_cat_links
```

```python
list_of_games=[]
links=[]
for cat in Dall_cat_links:
    if 'https://www.game' in cat:
        response=requests.get(cat)
        game_page=BeautifulSoup(response.text,'html.parser')
        game_link=game_page.find_all('div',class_='game-cell')
        for g in game_link:
            links=g.find('a',href=True)
            item=links['href']
            list_of_games.append(item)
            print(item)
list_of_games     
```

game meta_data

```python
details=[]
detail2=[]
likes=[]
rating=[]
r=[]
complete_list=[]

for game in list_of_games:
    try:
        page_of_game=requests.get(game)
        meta_data=BeautifulSoup(page_of_game.text,'html.parser') 

        url=game #url

        details=meta_data.find('div',class_='col-sm-8')

        date=details.find('span', class_='game-published')
        pub_date=date.text.split('with',1)[0][3:]  #publishing date
        plays=date.text.split(',',1)[1].split('h',1)[1].split("p",1)[0] #no of plays

        rating=details.find('div',class_='game-details-title')
        name=rating.find('h2')
        game_name=name.text.split(':',1)[0]#game

        likes=meta_data.find_all('span',class_='count')
        game_likes=likes[0].text#likes
        game_dislikes=likes[1].text#dislikes


        complete_list.append({'game':game_name, 
                              'url':url, 
                              'likes':game_likes, 
                              'dislikes':game_dislikes, 
                              'pub_date':pub_date,
                              'no of plays':plays})
        print({'game':game_name, 
                              'url':url, 
                              'likes':game_likes, 
                              'dislikes':game_dislikes, 
                              'pub_date':pub_date,
                              'no of plays':plays})
    except:
        print("failed" , game)
       


    
 
  
```

```python
df=pd.DataFrame(complete_list)
```
