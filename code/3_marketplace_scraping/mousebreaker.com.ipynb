```python
import pandas as pd
import os
import requests
import re
import time
import numpy as np
import subprocess
import random
from bs4 import BeautifulSoup
```

```python
response = requests.get("https://www.mousebreaker.com")
soup = BeautifulSoup(response.text, 'html.parser')
```

```python
outer_containers=[]
inner_containers=[]
gameurl=[]
gamelist = []
gameset = set()



categories_blocks = soup.find_all('a', href=True)
for block in categories_blocks:
    if '/games/' not in block['href']:
        continue
    try:
        cat_url = block['href']
        cat_name = block.find('p', class_='menu-title').text
    except:
        continue
    
    new_url = "https://www.mousebreaker.com" + cat_url
    page_no = 1
    while True:
        print(new_url+'?page=' + str(page_no))
        response = requests.get(new_url+'?page=' + str(page_no))
        soup = BeautifulSoup(response.text, 'html.parser')
        
        
        games_in_page = set()
        games = soup.find_all('a', class_='tile')
        for g in games:
            if g['href'].startswith('https://'):
                gameurl = g['href']
            try:
                gametitle = g.find('div', class_='tile-title').text
            except:
                continue
            if gameurl == []:
                continue
            gamelist.append({'game':gametitle, 'url':gameurl, 'category':cat_name})
#             print({'game':gametitle, 'url':gameurl, 'category':cat_name})
            games_in_page.add(gameurl)
#             print({'game':gametitle, 'url':gameurl, 'category':cat_name})
        if games_in_page - gameset == set():
            print(games_in_page)
            
            print(gameset)
            break
        gameset = gameset | games_in_page
        page_no+=1
    break
```

```python
complete_list = []
for game in gamelist:
    gameurl = game['url']
    print(gameurl)
    response = requests.get(gameurl)
    soup = BeautifulSoup(response.text, 'html.parser')
    
    like_perc = 'NOT_FOUND'
    dislike_perc = 'NOT_FOUND'
    try:
        like_perc = soup.find_all('span', class_='rate-text')[0].text
        dislike_perc = soup.find_all('span', class_='rate-text')[1].text
    except:
        pass
    desc = 'NOT_FOUND'
    try:
        desc = soup.find('div', class_='description').text
    except:
        pass

    

    complete_list.append({'game':game['game'], 
                          'url':game['url'], 
                          'category':game['category'], 
                          'like_perc':like_perc, 
                          'dislike_perc':dislike_perc, 
                          'desc':desc})
    print({'game':game['game'], 
                          'url':game['url'], 
                          'category':game['category'], 
                          'like_perc':like_perc, 
                          'dislike_perc':dislike_perc, 
                          'desc':desc})
    
    
# df = pd.DataFrame(complete_list)   
```

```python
df=pd.DataFrame(complete_list)
```

```python
df.to_csv('mousebreaker_game_info.csv',index=False)
```
