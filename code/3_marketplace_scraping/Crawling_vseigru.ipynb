```python
import googletrans

raw_data = []

url = "http://vseigru.net"
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')

for x in soup.find_all('a', href=True):
    if '.html' not in x['href']:
        continue #skip (not game category)
    try:
        cat_name = x['title']
    except:
        continue
    
    cat_url = x['href']
#     print(cat_name, cat_url)
#     try:
#         cat_name_en = translator.translate(cat_name, dest='en')
#     except:
#         continue
    raw_data.append({'cat_url': url+cat_url, 
                     'cat_name': cat_name
#                      'cat_name_en': str(cat_name_en).split(',')[0]
                    })
df_cat = pd.DataFrame(raw_data)
```

```python
df_cat
```

```python
games = []
for i, row in df_cat.iterrows():
    #get list of games for the category
    print("Scraping categoty", row['cat_name'])
    response = requests.get(row['cat_url'])
    soup = BeautifulSoup(response.text, 'html.parser')
    try:
        game_list =  soup.select('div[class*="games-list"]')[0]
    except:
        continue #no games here
    for x in game_list.find_all('a', class_='game'):
        try:
            games.append({'game':x['title'], 
                          'url':"http://vseigru.net"+x['href'], 
                          'category':row['cat_name']})
        except:
            pass #no game or missing info
df_games = pd.DataFrame(games)
df_games.to_csv("./vseigru_game_info.csv")
```
