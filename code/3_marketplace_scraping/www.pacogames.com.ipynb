```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import re
```

```python
response=requests.get('https://www.pacogames.com')
soup=BeautifulSoup(response.text,'html.parser')
```

```python
# cats=soup.find('ul',class_='m-main__item')
categories=[]
cats=soup.find_all('li',class_='m-main__item')
for cat in cats[1:]:
    temp=cat.find('a',href=True)
    category=temp['href']
    categories.append(category)
categories
```

fetch all games from a category page

```python
lst=[]
linkss=[]
all_games=[]

for cat in categories:
    url='https://www.pacogames.com'+cat
    res=requests.get(url)
    games=BeautifulSoup(res.text,'html.parser')
    
    gms=games.find('div',class_='b-category__content')
    linkss=gms.find_all('a',href=True)
    for x in linkss:
        all_games.append(x['href'])

all_games    
```

go to game pages and fetch meta data

```python
thumbs=[]
complete_list=[]
desc=[]
for game in all_games:
    try:
        new_url='https://www.pacogames.com'+game
        print(new_url)
        r=requests.get(new_url)
        meta=BeautifulSoup(r.text,'html.parser')
        game_name=game.split('/',1)[1].split('/')[1] #game_name
        game_url=new_url #game_url
        plays=meta.find('span',class_='b-annot__stats items__item')
        game_plays=plays.text.split(' ',1)[0] #game_plays

        thumbs=meta.find_all('span',class_='b-annot__icon-name')
        game_likes=thumbs[3].text #likes
        game_dislikes=thumbs[4].text #dislikes
        rate_perc=meta.find('span',class_='rating__value').text#rate percentage
        info=meta.find('div',id='gameinfo')
        desc=info.find_all('p')
        game_desc=desc[0].text # description
        game_pub=desc[1].text#publishing_date

        complete_list.append({'game':game_name, 
                              'url':game_url, 
                              'category':'NOT_FOUND', 
                              'like_perc':rate_perc, 
                              'game_plays':game_plays, 
                              'likes':game_likes,
                              'dislikes':game_dislikes,
                              'desc':game_desc,
                             'pub_date':game_pub})
        print({'game':game_name, 
                              'url':game_url, 
                              'category':'NOT_FOUND', 
                              'like_perc':rate_perc, 
                              'game_plays':game_plays, 
                              'likes':game_likes,
                              'dislikes':game_dislikes,
                              'desc':game_desc,
                             'pub_date':game_pub})
    except:
        print('link failed'+new_url)
```

```python
df=pd.DataFrame(complete_list)
```

```python
df.to_csv('pacogames_game_info',index=False)
```
