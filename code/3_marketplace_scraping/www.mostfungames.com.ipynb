```python
import pandas as pd
import requests
import re
from bs4 import BeautifulSoup
```

```python
response = requests.get("https://www.mostfungames.com")
soup = BeautifulSoup(response.text, 'html.parser')
```

find urls of all the categories

```python
category_list=[]
cats=soup.find('div', class_='menu')
urls=cats.find_all('a',href=True)

for url in urls:
    cat_url=url['href']
    category_list.append(cat_url)
    
category_list
```

get game page links from each category

```python
games=[]
lst=[]
lst_games=[]
for category in category_list:
    res = requests.get(category)
    single_cat = BeautifulSoup(res.text, 'html.parser')
    games=single_cat.find_all('div',class_='games')
    for g in games:
        lst=g.find_all('a',href=True)
        for l in lst:
            a_game=l['href']
            print(a_game)
            lst_games.append(a_game)
lst_games
```

get meta data from game pages

```python
import time
plays=[]
temp=[]
complete_list=[]

for game in lst_games:
#     time.sleep(5)
    try:
        respns = requests.get(game)
        meta_data = BeautifulSoup(respns.text, 'html.parser')
        game_url=game #url
        author=meta_data.find('p',class_='game_author')
        game_author=author.find('a',href=True).text #author
        plays=meta_data.find_all('span',class_='playcount border-radius')
        game_plays=plays[2].text #plays
        votes=meta_data.find('span',class_='update_votes')
        game_votes=votes.text #votes
        temp=meta_data.find_all('p')
        game_description=temp[2].text #description
        game_name=temp[0].text.split('-',1)[1] #name



        complete_list.append({'game':game_name, 
                              'url':game_url, 
                              'author':game_author, 
                              'plays':game_plays, 
                              'votes':game_votes, 
                              'desc':game_description})
        print({'game':game_name, 
                              'url':game_url, 
                              'author':game_author, 
                              'plays':game_plays, 
                              'votes':game_votes, 
                              'desc':game_description})
    except:
        print("failed" , game)
       
```

```python
df=pd.DataFrame(complete_list)
```
