```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import re
```

1.get soup object

```python
response=requests.get('https://www.games1.in/')
soup=BeautifulSoup(response.text,'html.parser')
```

2.Get list of all the categories

```python
cat_urls=[]
cts=soup.find_all('span',class_='round')
for c in cts:
    cgrs=soup.find_all('a',href=True)
for cgr in cgrs:
    if '(' in cgr.text:
        u=cgr['href']
        cat_urls.append(u)
    else:
        pass
cat_urls
```

3.create links for all the categories and games in each category

```python
games_list=[]
for cat_url in cat_urls:
    res=requests.get('https://www.games1.in/'+cat_url)
    soop = BeautifulSoup(res.text, 'html.parser')
    outer_ct=soop.find('div',id='content')
    inner_ct=outer_ct.find_all('div',class_='game round')
    for i in inner_ct:
        game=i.find('a',href=True)
        gameurl = game['href']
        games_list.append(gameurl)
        print(gameurl)
         
```

```python
games_list=[]
for cat_url in cat_urls:
    res=requests.get('https://www.games1.in/'+cat_url)
    soop = BeautifulSoup(res.text, 'html.parser')
    outer_ct=soop.find('div',id='content')
    inner_ct=outer_ct.find_all('div',class_='game round')
    for i in inner_ct:
        game=i.find('a',href=True)
        response=requests.get('https:'+game['href'])
        sup= BeautifulSoup(response.text, 'html.parser')
        game_page=sup.find_all('div',class_='game round')
        for g in game_page:
            gp_link=g.find('a',href=True)
            games_list.append(gp_link['href'])
            print(gp_link['href'])
            
        
         
```

```python
sml=[]
attr=[]
desc=[]
game_desc=[]
b=[]
dx=[]
cats=[]
ctgry=[]
complete_list=[]
category=[]
attr='NOT_FOUND'
votes='NOT_FOUND'
likes='NOT_FOUND'
category='NOT_FOUND'
for gp in games_list:
    r=requests.get('http:'+gp)
    page_meta=BeautifulSoup(r.text,'html.parser')
    div=page_meta.find('small')
    try:
        attr=div.find_all('b')
        votes=attr[0].text
        likes=attr[1].text
        detail=page_meta.find('div',id='gameinfo')
        desc=detail.text.split('How to play',1)[0]
        game=gp.split('.in/',1)[1]
        url='http:'+gp
        cats=detail.find_all('span',class_='round')
        for cat in cats:
            ctgry=cat.find('a',href=True).text
    #         category.append(ctgry)
            category=ctgry
    except:
        print('values not found')
        
    complete_list.append({'game':game, 'url':url, 'category':category, 'votes':votes, 'like_perc':likes,  'desc':desc})

    print({'game':game, 'url':url, 'category':category, 'votes':votes, 'like_perc':likes,  'desc':desc})


```
