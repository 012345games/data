```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import re
```

```python
response=requests.get('https://yourgames.io')
soup=BeautifulSoup(response.text,'html.parser')
```

```python
cats=[]
cat_links=[]
temp=[]
categories=[]
cats=soup.find_all('div',class_='category')
for cat in cats:
    cat_links=cat.find_all('a',href=True)
    temp1=cat_links[0]['href']
#   temp2=cat_links[1]['href']
    categories.append(temp1)
    
       
```

fetch category pages to collect game urls

```python
game_urls=[]
temp=[]
game_links=[]
game_names=[]
for category in categories:
    new_url='https://yourgames.io'+category
    res=requests.get(new_url)
    games=BeautifulSoup(res.text,'html.parser')
    game_urls=games.find('div',class_='games-container')
    temp=game_urls.find_all('a',href=True)
    for item in temp:
        game_link=item['href']
        name=item.find('div',class_='filter')
        game_name=name.text
        game_links.append(game_link)
        game_names.append(game_name)
    break
print(game_names)
print(game_links)
```

use game page links to fetch games and collect meta data

```python
rates=[]
details=[]
complete_list=[]
game_release_date='NOT_FOUND'
for link in game_links:
    try:
        full_url='https://yourgames.io'+link
        r=requests.get(full_url)
        meta_data=BeautifulSoup(r.text,'html.parser')
        rates=meta_data.find_all('p',class_='rate-btn-txt')
        game_likes=rates[0].text  #likes
        game_dislikes=rates[1].text #dislikes
        attr=meta_data.find('div',class_='game-description')
        details=attr.find_all('p')
        game_desc=details[0].text #desc
        game_release_date=details[1].text #rel_date
        game_developer=details[2].text #devloper
        game_name=link

        complete_list.append({'game':game_name, 
                              'url':full_url, 
                              'likes':game_likes, 
                              'dislikes':game_dislikes, 
                              'desc':game_desc,
                              'release_date':game_release_date,
                              'developer':game_developer})
        print({'game':game_name, 
                              'url':full_url, 
                              'likes':game_likes, 
                              'dislikes':game_dislikes, 
                              'desc':game_desc,
                              'release_date':game_release_date,
                              'developer':game_developer})
    except:
        print('couldnot resolve name' + link)

    

        
```

```python
df=pd.DataFrame(complete_list)
```
