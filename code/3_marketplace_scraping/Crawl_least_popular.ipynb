### www.fukgames.com

```python
page_no = 1
raw_data = []

URL = "https://www.fukgames.com"

raw_data = []

while page_no<465:
    if page_no % 50 == 0:
        print("page_no", page_no)
    response = requests.get(URL+'/'+str(page_no))
    soup = BeautifulSoup(response.text, 'html.parser')
    
    for x in soup.find_all('td', align="center"):
        game_url = x.find('a', href=True)['href']
        if "//www.fukgames" not in game_url:
            continue
        game_title = x.find('span').text
        raw_data.append({'game':game_title.split(' (')[0], 
                         'url':game_url.split("//")[1], 
                         'game_rate':game_title.split(' (')[1].split(')')[0]})
    page_no+=1


pd.DataFrame(raw_data).to_csv("fukgames_game_info.csv")
```

### alliogames.net

```python
URL = "http://alliogames.net"

#task 1 : get urls of game categories
response = requests.get(URL)
soup = BeautifulSoup(response.text, 'html.parser')
```

```python
soup
```

```python
# soup.find_all('ul', class_='sub-menu')
raw_data= []

cat_menu = soup.find('ul', class_='sub-menu')
for x in cat_menu.find_all('li'):
    print(x)
    raw_data.append({'category': x.text[:-1], 'category_url':x.find('a')['href']})
df_cat = pd.DataFrame(raw_data)

df_cat
```

```python
#step 2: iterate on categories and pages

raw_data_game = []

for i, row in df_cat.iterrows():
    print(i, row['category'], row['category_url'])
    
    page_no = 1
    while True:
        url = row['category_url'] + '/page/' + str(page_no)
        
        print('Requesting url=', url, "page=", page_no)
        response = requests.get(url)
        
        #if final url equal to home page, then stop
        if response.url == 'http://www.alliogames.net':
            print("stopping because no more pages")
            break
        
        soup = BeautifulSoup(response.text, 'html.parser')
        game_containers = soup.select('div[class*="thumb post"]')
      
        for x in game_containers:
            game_url = x.find('a')['href']
            game_name = game_url.split('/')[-1]
            raw_data_game.append({'game': game_name, 'url':game_url, 'category':row['category']})
        
        page_no+=1            

df_games = pd.DataFrame(raw_data_game)
```

```python
df_games.to_csv("alliogames_game_info.csv")
```

### gamepix.com

```python
URL = "https://www.gamepix.com/games-catalogue"

#get games catalogue
page_no = 1
raw_data = []
while True:
    response = requests.get(URL+'/'+str(page_no))
    print("Scraping", URL+'/'+str(page_no))
    soup = BeautifulSoup(response.text, 'html.parser')
    game_containers = soup.select('div[class*="GameTile"]')
    if game_containers==[]:
        break
    for x in game_containers:
#         print(x.find('a')['href'])
#         print(x.find('h3', class_='card-title').text)
        raw_data.append({'game':x.find('h3', class_='card-title').text, 
                         'gameurl': x.find('a')['href']})
    page_no+=1
```

```python
df_games = pd.DataFrame(raw_data)
df_games.head()
```

```python
raw_data = []

#get other game metadata
for i,row in df_games.iterrows():
    url = "https://www.gamepix.com" + row['gameurl']
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')
    
    #find likes/dislikes
    rates = [int(x.text[1:]) for x in soup.select('div[class*="Vote vote-container"]')[0].find_all('p')]
    likes = rates[0]
    dislikes = rates[1]
    
    print("processing", url)
    
    #author
    author = soup.find("p", class_='game-studio').text
    
    #desc
    desc_container = soup.find_all('article')[0].find_all('div')
    desc = desc_container[3].text
    
    #category
    try:
        cat = desc_container[4].text.split("belongs to ")[1].split(" and")[0]
    except:
        cat = 'NOT_FOUND'
    
    #score
    try:
        score = float(desc_container[4].text.split("score of ")[1].split('.')[0])
    except:
        score = 'NOT_FOUND'
    
    #published date
    try:
        published = desc_container[4].text.split("published on ")[1].split(' ')[0]
    except:
        published = 'NOT_FOUND'
    
    #update date
    try:
        updated = desc_container[4].text.split("updated on ")[1].split('.')[0]
    except:
        updated = 'NOT_FOUND'
    
    #apple store
    apple_store_url = 'NOT_FOUND'
    if soup.find('a', class_='app-store-badge')!=None:
        apple_store_url = soup.find('a', class_='app-store-badge')['href']
    #play store
    play_store_url = 'NOT_FOUND'
    if soup.find('a', class_='play-store-badge')!=None:
        play_store_url = soup.find('a', class_='play-store-badge')['href']
    
    raw_data.append({'game':row['game'], 
                     'url': "https://www.gamepix.com" + row['gameurl'], 
                     'category': cat, 
                     'likes': likes, 
                     'dislikes':dislikes, 
                     'published_on':published, 
                     'updated_on':updated, 
                     'author': author, 
                     'score':score, 
                     'desc':desc})
    
#     print({'game':row['game'], 
#                      'gameurl': row['gameurl'], 
#                      'category': cat, 
#                      'likes': likes, 
#                      'dislikes':dislikes, 
#                      'published_on':published, 
#                      'updated_on':updated, 
#                      'author': author, 
#                      'score':score, 
#                      'desc':desc})

df_game_info = pd.DataFrame(raw_data)
df_game_info
```

```python
df_game_info.to_csv("gamepix_game_info.csv")
```

### jogosjogos2.com

```python
page_no = 1
raw_data = []

URL = "https://www.jogosjogos2.com"

raw_data = []

while True: 
    print("page_no", page_no)
    response = requests.get(URL+'/'+str(page_no))
    soup = BeautifulSoup(response.text, 'html.parser')
    
    for x in soup.find_all('td', align="center"):
        game_url = x.find('a', href=True)['href']
        if "//www.jogosjogos2" not in game_url:
            continue
        game_title = x.find('span').text
        
        game_rating = x.find('span').text
        
        raw_data.append({'game':game_title.split(' (')[0], 
                         'url':game_url.split("//")[1], 
                         'game_rate':game_title.split(' (')[1].split(')')[0]})
    page_no+=1
df = pd.DataFrame(raw_data)
```

```python
# df['url'] = 'https://' + df['url']
df
```

```python
raw_data = []

processed = set()
all_games = set(df['game'].unique().tolist())
```

```python
while True:
    for i, row in df.iterrows():
        if all_games - processed == set():
            #all games
            break
        
        if row['game'] in processed:
            continue
        
        url = row['url'].replace(' ', '')
        print(i, url)
        
        try:
            response = requests.get(url)
        except:
            print('non 200, sleeping')
            time.sleep(10)
            continue


        soup = BeautifulSoup(response.text, 'html.parser')
        info = soup.find('div', class_='gray')
        
        
        try:
            playcount  = int(info.text.split('played ')[1].split(' times')[0])
        except:
            playcount = 'NOT_FOUND'
            
        try:
            tags = [x.text for x in info.find_all('a')]
        except:
            tags = []
        
        try:
            desc = soup.find('p', class_='description').text.split('\n')[0]
        except:
            desc = 'NOT_FOUND'

    #     print(row['url'], row['game_rate'], playcount, tags, desc)
        raw_data.append({'game':row['game'], 
                         'url':url, 
                         'playcount':playcount, 
                         'tags':tags, 
                         'desc':desc})
        processed.add(row['game'])
```

```python
df_gameinfo = pd.DataFrame(raw_data)



def get_rate(x):
    try:
        return df[df['game']==x]['game_rate'].values[0]
    except:
        return 'NOT_FOUND'
df_gameinfo['rate'] = df_gameinfo['game'].apply(get_rate) 

df_gameinfo.to_csv("jogosjogos2_game_info.csv")
df_gameinfo
```
