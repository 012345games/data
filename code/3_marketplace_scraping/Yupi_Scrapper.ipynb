# yupi.io

```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import re
```

```python
response = requests.get("https://www.yupi.io")
soup = BeautifulSoup(response.text, 'html.parser')
```

```python
cat_urls=[]

categories = soup.find_all('a', href=True) #finding all category urls
for category in categories:
    if '/c/' not in category['href']:
        continue
    try:
        cat_url=category['href']
        cat_urls.append(cat_url)
    except:
        continue

cat_urls

```

```python
link_all_games=[]
for cat_page in cat_urls:
        games_in_cat = requests.get(cat_page)
        soup = BeautifulSoup(games_in_cat.text, 'html.parser')
        all_games = soup.find_all('a', class_='game')
        for one_game in all_games:
                    link=one_game['href']
                    link_all_games.append(link)
        
link_all_games 
```

```python
len(link_all_games)
```

```python
aa=[]
bb=[]
cc=[]
ss=[]
desc=[]
plays=[]
complete_list=[]


for link in link_all_games:
    game_url=link
    f=link.split('b',1)[0]
    s=link.split('o',1)[1]
    game_page = requests.get(f+'httpplay'+s)
    game_page_soup= BeautifulSoup(game_page.text, 'html.parser')
    
    plays=game_page_soup.find_all('div',class_='gameinfo')
    for p in plays:
        pla=p.find('b').text
        number= pla
        noofplays=number
        
    aa=game_page_soup.find_all('span',id='liked')
    for a in aa:
        bb=a.find('a')
        k = bb['onclick']
        nooflikes=k
        
    xx=game_page_soup.find_all('span',id='disliked')
    for x in xx:
        yy=a.find('a')
        z = yy['onclick']
    noofdislikes=z
    
    try:
        desc = game_page_soup.find_all('div', class_='game-description')
        for d in desc:
            ss=d.find('p').getText()
            description=ss 
    except:
        pass

   
    
    complete_list.append({'game':game_url, 
                                     'playcount':noofplays, 
                                      'like_perc':nooflikes, 
                                      'dislike_perc':noofdislikes, 
                                      'desc':description})
    print({'game':game_url, 
                                      'playcount':noofplays, 
                                      'like_perc':nooflikes, 
                                      'dislike_perc':noofdislikes, 
                                      'desc':description})
  

```

```python
df=pd.DataFrame(complete_list)
```
