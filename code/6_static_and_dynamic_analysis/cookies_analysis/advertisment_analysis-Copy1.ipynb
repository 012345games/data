```python
# https://github.com/uBlock-LLC/uBlock/issues/1406
file1 = open("output_files/simple_ad.txt", 'r')
Lines = file1.readlines()
Lines
lst = []
for l in listy:
    for L in Lines:
        if l in L:
#             print(l)
            lst.append(l)
        else:
            pass

from collections import Counter
c = Counter(lst)
c
df_ht = pd.DataFrame(c, index=['count']).transpose()
df_ht.index.rename('ad domain', inplace=True)
df_ht.reset_index(inplace=True)
df_ht
df_ht = df_ht.sort_values(by = "count",ascending = False)
df_ht
sm = df_ht['count'].sum()
l = []
for index, row in df_ht.iterrows():
#     print(row['count'])
#     print(sm)
#     print((row['count']/sm)*100)
    l.append((row['count']/sm)*100)
df_ht['perc'] = l
df_ht
```
