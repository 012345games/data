# Extracting Games' Marketplace data from SimilarWeb

```python
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.support.ui import WebDriverWait
import time

def get_info(html_source_code):
    
    _Soup = BeautifulSoup(html_source_code, 'html.parser')

    # Web game marketplace info.

    wgm_place_info = {}

    # Company info
    company_info = {}

    ci_elements = _Soup.findAll("dd",{"class":"data-company-info__list-item data-company-info__list-item--value"})
    
    try:
        company_info['Company'] = ci_elements[0].text.strip()
        company_info['Year_Founded'] = ci_elements[1].text.strip()
        company_info['Employees'] = ci_elements[2].text.strip()
        company_info['HQ'] = ci_elements[3].text.strip()
        company_info['Annual_Revenue'] = ci_elements[4].text.strip()
        company_info['Industry'] = ci_elements[5].text.strip()
    except:
        company_info['Company'] = "NA"
        company_info['Year_Founded'] = "NA"
        company_info['Employees'] = "NA"
        company_info['HQ'] = "NA"
        company_info['Annual_Revenue'] = "NA"
        company_info['Industry'] = "NA"


    wgm_place_info["CompanyInfo"] = company_info


    # Traffic & Engagement Last Month
    te_elements = _Soup.findAll("span",{"class":"wa-traffic__engagement-item-value"})

    te_info = {}
    try:
        te_info["Total_Visits"] = te_elements[0].text.strip()
    except:
        # Similarweb does not present data for websites with < 5K estimated visits for the last month. 
        # Only data verified by Google Analytics will be displayed.
        te_info["Total_Visits"] = "NA"
    try:
        te_info["Last_Month_Change"] = te_elements[1].text.strip()
    except:
        te_info["Last_Month_Change"] = "NA"
    try:
        te_info["Avg_Visit_Duration"] = te_elements[2].text.strip()
    except:
        te_info["Avg_Visit_Duration"] = "NA"
    try:
        te_info["Bounce_Rate"] = te_elements[3].text.strip()
    except:
        te_info["Bounce_Rate"] = "NA"
    try:
        te_info["Pages_per_Visit"] = te_elements[4].text.strip()
    except:
        te_info["Pages_per_Visit"] = "NA"

    wgm_place_info["TrafficEngagement"] = te_info

    # Demographic Info
    demographic_info = {}

    try:
        demographic_info["Female"] = _Soup.findAll("span",{"class":"wa-demographics__gender-legend-item-value"})[0].text
    except:
        demographic_info["Female"] = "NA"
    try:
        demographic_info["Male"] = _Soup.findAll("span",{"class":"wa-demographics__gender-legend-item-value"})[1].text
    except:
        demographic_info["Male"] = "NA"
    wgm_place_info["DemographicInfo"] = demographic_info

    # Demographics Age Info

    demographic_age_info = {}

    try:
        demographic_age_info["18-24"] = _Soup.findAll("tspan",{"class":"wa-demographics__age-data-label"})[0].text.strip()
    except:
        demographic_age_info["18-24"] = "NA"
    try:
        demographic_age_info["25-34"] = _Soup.findAll("tspan",{"class":"wa-demographics__age-data-label"})[1].text.strip()
    except:
        demographic_age_info["25-34"] = "NA"
    try:
        demographic_age_info["35-44"] = _Soup.findAll("tspan",{"class":"wa-demographics__age-data-label"})[2].text.strip()
    except:
        demographic_age_info["35-44"] = "NA"
    try:
        demographic_age_info["45-54"] = _Soup.findAll("tspan",{"class":"wa-demographics__age-data-label"})[3].text.strip()
    except:
        demographic_age_info["45-54"] = "NA"
    try:
        demographic_age_info["55-64"] = _Soup.findAll("tspan",{"class":"wa-demographics__age-data-label"})[4].text.strip()
    except:
        demographic_age_info["55-64"] = "NA"
    try:
        demographic_age_info["65+"] = _Soup.findAll("tspan",{"class":"wa-demographics__age-data-label"})[5].text.strip()
    except:
        demographic_age_info["65+"] = "NA"

    wgm_place_info["DemographicAgeInfo"] = demographic_age_info
    
    return wgm_place_info
    
fin = [x.strip() for x in open('/home/muhammad/work/Projects/MalwareTriage/dev/dike/VTScan/CODS2023/1248_websites.csv').readlines()[1:]]

webgame_info = {}

count = 0
for website in fin[count:]:
    # initialize the Chrome driver
    driver = webdriver.Chrome("/home/muhammad/work/Projects/MalwareTriage/dev/dike/VTScan/CODS2023/chromedriver")

    # head to EasyChair login page
    driver.get("https://www.similarweb.com/website/%s/"%website)
    
    time.sleep(15)
    
    html_source_code = driver.execute_script("return document.body.innerHTML;")
    
    _wgm_place_info = get_info(html_source_code)
    
    webgame_info[website] = _wgm_place_info   
    
    print("%s :: %s :: %s" % (count, website, _wgm_place_info))
    
    driver.quit()
    count+=1 
```
