### Get all IPs associated with marketplace url

*For smaller sites, I would do a WHOIS on the domain, then do a NS query on each DNS server listed to get a complete list of all possible DNS servers for that domain. Finally, query each DNS server for the FQDN that you want the IP list for*

```python
raw_data = []
for mp in MARKETPLACES:
    #try to get the nameservers from whois
    try:
        domain = whois_ip(mp)
        name_servers = list(domain.__dict__['name_servers'])
    except:
#         name_servers = []
        pass
    
    ips_ns = ['']
    
    #add all ips of name servers
    for ns in name_servers:
        p = subprocess.Popen(['nslookup', ns], stdout=subprocess.PIPE, stderr=subprocess.PIPE)
        out, err = p.communicate()
        for line in out.decode('utf-8').split('\n'):
            if "Address: " in line and len(line.split("Address: ")[1])<16:
                ips_ns.append(line.split("Address: ")[1])
    
    for ip_ns in ips_ns:
        p = subprocess.Popen(['nslookup', mp, ip_ns], stdout=subprocess.PIPE, stderr=subprocess.PIPE)
        out, err = p.communicate()
        for line in out.decode('utf-8').split('\n'):
            if "Address: " in line and len(line.split("Address: ")[1])<16:
                raw_data.append({'website':mp, 'ip':line.split("Address: ")[1]})
    print("Done", mp)
```

```python
import pandas as pd
import subprocess

raw_data = []

for i, row in df.iterrows():
    print("processed", row['website'])
    #run ipgeolocation.py
    p = subprocess.Popen(['python3', './../dynamic_analysis/IPGeoLocation/ipgeolocation.py', '-t', row['ip']], 
                         stdout=subprocess.PIPE, stderr=subprocess.PIPE)
    out, err = p.communicate()
    
    #parse output
    city = isp = country_code = asn = None
    for line in out.decode('utf-8').split('\n'):
        if "Country Code" in line:
            country_code = line.split("Country Code: ")[1]
        if "City" in line:
            city = line.split("City: ")[1]
        if "ISP" in line:
            isp = line.split("ISP: ")[1]
        if "ASN" in line:
            asn = line.split("ASN: ")[1].split()[0]
    try:
        raw_data.append({'website':row['website'], 
                         'ip':row['ip'], 
                         'country_code':country_code.split('32m')[1].split('\x1b')[0], 
                         'city':city.split('32m')[1].split('\x1b')[0], 
                         'isp':isp.split('32m')[1].split('\x1b')[0], 
                         'asn':asn.split('32m')[1].split('\x1b')[0]})
    except:
        raw_data.append({'website':row['website'], 
                         'ip':row['ip'], 
                         'country_code':country_code, 
                         'city':city, 
                         'isp':isp, 
                         'asn':asn})
        
    time.sleep(0.5)
    
df_geoloc = pd.DataFrame(raw_data)
```

### Analyse IPs with VT

```python
import requests, json, sqlite3, time

def get_whois_data(text, whois_ts):
    try:
        temp = text.split("\n")
    except:
        temp = "NotApplicable"
    try:
        domain = temp[0].split(":")[-1].strip()
    except:
        domain = "NotApplicable"
    try:
        nserver1 = temp[1].split(":")[-1].strip()
    except:
        nserver1 = "NotApplicable"
    try:
        nserver2 = temp[2].split(":")[-1].strip()
    except:
        nserver2 = "NotApplicable"
    try:
        nstate = temp[3].split(":")[-1].strip()
    except:
        nstate = "NotApplicable"
    try:
        person = temp[4].split(":")[-1].strip()
    except:
        person = "NotApplicable"
    try:
        registrar = temp[5].split(":")[-1].strip()
    except:
        registrar = "NotApplicable"
    try:
        admin_contact = temp[6].split(":")[-1].strip()
    except:
        admin_contact = "NotApplicable"
    try:
        created_at = temp[7].split(":")[-1].strip()
    except: 
        created_at = "NotApplicable"
    try:
        paid_till = temp[8].split(":")[-1].strip()
    except:
        paid_till = "NotApplicable"
    try:
        source = temp[8].split(":")[-1].strip()
    except:
        source = "NotApplicable"
    return whois_ts, domain, nserver1, nserver2, nstate,person, registrar, admin_contact, created_at, paid_till, source

def get_detected_downloaded_samples(item, domain):     
    return item['date'], domain, item['positives'], item['total'], item['sha256']

def get_undetected_downloaded_samples(item, domain):    
    return item['date'], domain, item['positives'], item['total'], item['sha256']

def get_detected_urls(item, domain):    
    return item['scan_date'], domain, item['positives'], item['total'], item['url']

def get_ip_resolve(item, domain):
    return item['last_resolved'], domain, item['hostname']

def get_undetected_referrers(item, domain):
    return domain, item['positives'], item['total'], item['sha256']

def get_detected_referrers(item, domain):
    return domain, item['positives'], item['total'], item['sha256']

def detected_communicating_samples(item, domain):    
    return item['date'], domain, item['positives'], item['total'], item['sha256']

def undetected_communicating_samples(item, domain):    
    return item['date'], domain, item['positives'], item['total'], item['sha256']

conn = sqlite3.connect('./marketplace_IPs_VT.db')
cc = conn.cursor()
try:
    cc.execute('''CREATE TABLE whois_info (whois_ts TEXT, IP TEXT, nserver1 TEXT, nserver2 TEXT, nstate TEXT,person TEXT, registrar TEXT, admin_contact TEXT, created_at TEXT, paid_till TEXT, source TEXT)''') 
    cc.execute('''CREATE TABLE whois (whois_ts TEXT, IP TEXT, whois_text TEXT)''')  
    cc.execute('''CREATE TABLE detected_downloaded_samples (detection_ts TEXT, IP TEXT, npositives TEXT, nantivirustools TEXT, sha256 TEXT)''') 
    cc.execute('''CREATE TABLE undetected_downloaded_samples (detection_ts TEXT, IP TEXT, npositives TEXT, nantivirustools TEXT, sha256 TEXT)''') 
    cc.execute('''CREATE TABLE IP_info (IP TEXT, network TEXT, asn TEXT, as_owner TEXT, country TEXT, response_code TEXT, verbose_msg TEXT)''') 
    cc.execute('''CREATE TABLE detected_urls (scan_date TEXT, IP TEXT, npositives TEXT, nantivirustools TEXT, url TEXT)''')
    cc.execute('''CREATE TABLE ip_resolve (resolve_ts TEXT, IP TEXT, resolved_to_ip TEXT)''')
    cc.execute('''CREATE TABLE detected_referrers(IP TEXT, npositives TEXT, nantivirustools TEXT, sha256 TEXT )''')
    cc.execute('''CREATE TABLE undetected_referrers(IP TEXT, npositives TEXT, nantivirustools TEXT, sha256 TEXT )''')
    cc.execute('''CREATE TABLE detected_communicating_samples (detection_ts TEXT, IP TEXT, npositives TEXT, nantivirustools TEXT, sha256 TEXT)''') 
    cc.execute('''CREATE TABLE undetected_communicating_samples (detection_ts TEXT, IP TEXT, npositives TEXT, nantivirustools TEXT, sha256 TEXT)''') 
except:
    pass

IPs = df['ip'].values.tolist()

headers = { "Accept-Encoding": "gzip, deflate", "User-Agent" : "gzip,  Gio Tan" }

API_KEY='2a7e38da1ca3f07faad92d11c37ef70a25e768aa6d86dbb27d7e1720921b0e65' #20K per day
url = 'https://www.virustotal.com/vtapi/v2/ip-address/report'

count = 0
for ip in IPs:
    
    params = {'apikey':API_KEY,'ip':ip} 

    response = requests.get(url, params=params)

    response_dict = response.json()
    try:
        whois_timestamp = response_dict['whois_timestamp']
    except:
        whois_timestamp = "NotApplicable"
    try:
        country = response_dict['country']
    except:
        country = "NotApplicable"
    try:
        network = response_dict['network']
    except:
        network = "NotApplicable"
    try:
        response_code = response_dict['response_code']
    except:
        response_code = "NotApplicable"
    try:
        as_owner = response_dict['as_owner']
    except:
        as_owner = "NotApplicable"
    try:
        verbose_msg = response_dict['verbose_msg']
    except:
        verbose_msg = "NotApplicable"
    try:
        asn = response_dict['asn']
    except:
        asn = "NotApplicable"
    
    try:
        cc.execute("INSERT INTO IP_info VALUES (?,?,?,?,?,?,?)", (ip, network, asn, as_owner, country, response_code,verbose_msg)) 
        conn.commit()
    except:
        pass
    
    try:
        # some VT lookups do not have whois data
        if 'whois_timestamp' in response_dict.keys() and 'whois' in response_dict.keys():
            cc.execute("INSERT INTO whois VALUES(?,?,?)", (whois_timestamp,ip, response_dict['whois']))
            cc.execute("INSERT INTO whois_info VALUES(?,?,?,?,?,?,?,?,?,?,?)", (get_whois_data(response_dict['whois'], whois_timestamp)))
            conn.commit

        if 'undetected_downloaded_samples' in response_dict.keys():
            for item in response_dict['undetected_downloaded_samples']:
                cc.execute("INSERT INTO undetected_downloaded_samples VALUES (?,?,?,?,?)", (get_undetected_downloaded_samples(item, ip)))
                conn.commit()
            for item in response_dict['detected_downloaded_samples']:
                cc.execute("INSERT INTO detected_downloaded_samples VALUES (?,?,?,?,?)", (get_detected_downloaded_samples(item, ip))) 
                conn.commit()
        elif 'detected_communicating_samples' in response_dict.keys():
            for item in response_dict['detected_communicating_samples']:
                cc.execute("INSERT INTO detected_communicating_samples VALUES (?,?,?,?,?)", (get_detected_downloaded_samples(item, ip))) 
                conn.commit()
            for item in response_dict['undetected_communicating_samples']:
                cc.execute("INSERT INTO undetected_communicating_samples VALUES (?,?,?,?,?)", (get_detected_downloaded_samples(item, ip)))
                conn.commit()


        if u'detected_urls' in response_dict.keys():
            for item in response_dict[u'detected_urls']:
                cc.execute("INSERT INTO detected_urls VALUES (?,?,?,?,?)", (get_detected_urls(item, ip)))
                conn.commit()

        if u'resolutions' in response_dict.keys():
            for item in response_dict[u'resolutions']:
                #print get_ip_resolve(domain, item)    
                cc.execute("INSERT INTO ip_resolve VALUES (?,?,?)", (get_ip_resolve(item, ip)))
                conn.commit()
        if u'undetected_referrer_samples' in response_dict.keys():
            for item in response_dict[u'undetected_referrer_samples']:
                cc.execute('''INSERT INTO undetected_referrers VALUES(?,?,?,?)''',(get_undetected_referrers(item, ip)))
                conn.commit()
            for item in response_dict[u'detected_referrer_samples']:
                cc.execute('''INSERT INTO detected_referrers VALUES(?,?,?,?)''',(get_detected_referrers(item, ip)))
                conn.commit()

        elif u'detected_communicating_samples' in response_dict.keys():
            for item in response_dict[u'undetected_communicating_samples']:
                cc.execute('''INSERT INTO undetected_communicating_samples VALUES(?,?,?,?)''',(undetected_communicating_samples(item, ip)))
                conn.commit()
            for item in response_dict[u'detected_communicating_samples']:
                cc.execute('''INSERT INTO detected_communicating_samples VALUES(?,?,?,?)''',(detected_communicating_samples(item, ip)))
                conn.commit()
    except:
        pass
    
    print ("%s:: %s"% (count, ip))
    count += 1
conn.close()
```
