### Example: all marketplaces related to https://www.fukgames.com/ are similar

```python
# MP_EX = ["www.fukgames.com", ""]

res = requests.get("https://www.fukgames.com")
soup = BeautifulSoup(res.text, 'html.parser')

MP_EX  = [x['href'] for x in soup.find('div', id='links').find_all('a', href=True)]

htmls = {}
for x in MP_EX:
    print(x)
    try:
        htmls[x] = requests.get(x).text
    except:
        htmls[x] = ''
```

```python
for x in htmls:
    for y in htmls: 
        if x==y or htmls[x]=='' or htmls[y]=='':
            continue
        sim = similarity(htmls[x], htmls[y])
        if sim>0.1:
            print(x, 'vs', y)
            print(similarity(htmls[x], htmls[y]), style_similarity(htmls[x], htmls[y]), structural_similarity(htmls[x], htmls[y]))
```

```python
#get all html
MARKETPLACES = []
with open("marketplaces_xyz.csv") as fin: #list of 1248 websites
    for line in fin:
        MARKETPLACES.append(line.split(',')[0])

#get html(s)
htmls = {}
for mp in MARKETPLACES:
    print("Retrieving html for", mp)
    try:
        html = requests.get("http://" + mp, timeout=10).text
    except:
        continue
    with open("./marketplace_htmls/" + mp + "__html.text", 'w') as fout:
        fout.write(html)       
```

```python
#compute similarity
raw_data = []
for mpX in MARKETPLACES:
    print("Processing", mpX)
    try:
        with open("./marketplace_htmls/" + mpX + "__html.text") as fin:
            htmlX = fin.read()
    except:
        continue
    
    for mpY in MARKETPLACES:
        if mpX == mpY:
            continue
        try:
            with open("./marketplace_htmls/" + mpY + "__html.text") as fin:
                htmlY = fin.read()
        except:
            continue
        try:
            raw_data.append({'websiteX': mpX, 
                             'websiteY': mpY, 
                             'sim': similarity(htmlX, htmlY), 
                             'struct_sim': structural_similarity(htmlX, htmlY), 
                             'style_sim': style_similarity(htmlX, htmlY)})
        except:
            pass

        
df = pd.DataFrame(raw_data)
df.to_csv("mp_html_sim.csv")
```
