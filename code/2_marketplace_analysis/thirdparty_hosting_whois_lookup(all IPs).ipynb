```python
import pandas as pd
import socket
```

```python
# using "socket.gethostbyname_ex()" to get all of the ipv4 addresses registered with a marketplace
df = pd.read_csv("mp_1248.csv", header = None)
lst_1248 = df[0].to_list()

L = []
lst_error = []

for mp in lst_1248:
    try:
        lookup_data = socket.gethostbyname_ex(mp)
        print(mp, lookup_data[2])
        d = {"mp":mp,"#ips":len(lookup_data[2]),"addresses":lookup_data[2]}
        L.append(d)
    except:
        lst_error.append(mp)
        print("Error in :", mp)
pd.DataFrame(L).to_csv("all_ips_per_hostname.csv",index = False)
```

```python
# doing whois ip for all of the ips found against one hostname
import whois
from ipwhois import IPWhois
from pprint import pprint
import ast
import time

df = pd.read_csv("all_ips_per_hostname.csv")
L1 = []
L2= []
df = df[:300]
for index, row in df.iterrows():
    print(row[0],row[2])
    # whois information of hostname
    domain= row[0]
    whois.whois(domain)
    
    # whois information of ipv4 and/or ipv6 addresses
    results = []
    for item in ast.literal_eval(row[2]):
        
        print(item)
        obj = IPWhois(item)
        results.append(obj.lookup_rdap())
        
        d1 = {"ip":item,"whois":results}
        L1.append(d1)
        time.sleep(10)
        
#         pprint(results)
    d2= {"mp":domain,"whois_hostname":whois.whois(domain),"whois_ips":L1}
    L2.append(d2)

pd.DataFrame(L2).to_csv("whois_all_ip_per_mp_0_300.csv",index = False)
```
